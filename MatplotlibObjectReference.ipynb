```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 1, 10)
y = x ** 2
plt.plot(x, y)
```

```python
fig = plt.figure()
axes = fig.add_axes([0.1, 0.1, 0.8, 0.8])

axes.plot(x, y)
axes.set_xlabel("X")
axes.set_ylabel("Y")
axes.set_title("X vs Y")
plt.show()
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1, 0.1, 0.8, 0.8])
axes2 = fig.add_axes([0.6, 0.18, 0.25, 0.2]) # -> left, _ bottom, <- right, - up
axes3 = fig.add_axes([0.15, 0.76, 0.3, 0.1])
axes2.plot(x, y)
axes2.set_xlabel("X")
axes2.set_ylabel("Y")
axes2.set_title("X vs Y")

axes3.plot(x, y)
axes3.set_xlabel("Y")
axes3.set_ylabel("X")
axes3.set_title("X vs Y")
plt.show()
```
